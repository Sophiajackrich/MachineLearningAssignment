### Model Preparation

```python
models = []
models.append(('LR', LinearRegression()))
models.append(('KR', KNeighborsRegressor()))
models.append(('SVR', SVR()))
models.append(('DT', DecisionTreeRegressor()))
models.append(('RF', RandomForestRegressor()))
models.append(('BR', BaggingRegressor()))
models.append(('AR', AdaBoostRegressor()))
models.append(('ER', ExtraTreesRegressor()))
models.append(('GR', GradientBoostingRegressor()))
```

### Training Dataset

```python
results = []
names = []
scoring = 'neg_mean_squared_error'
for name, model in models:
    kfold = model_selection.KFold(n_splits= 10, random_state = 30)
    mse_cv = model_selection.cross_val_score(model, X_train, y_train, cv = kfold, scoring = scoring)
    results.append(mse_cv)
    names.append(name)
    mse ='%s: %.2f, %.2f, %.2f'% (name, mse_cv.mean(), mse_cv.var(), mse_cv.std())
    print(mse)
```

### Plotting Algorithm Comparison in a Boxplot

```python
fig = plt.figure()
fig.suptitle("Algorithm comparison and the training dataset")
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

### Model Evaluation on Testing Dataset

```python
results = []
names = []
scoring = 'neg_mean_squared_error'
for name, model in models:
    kfold = model_selection.KFold(n_splits= 10, random_state = 30)
    mse_cv = model_selection.cross_val_score(model, X_test, y_test, cv = kfold, scoring = scoring)
    results.append(mse_cv)
    names.append(name)
    mse ='%s: %.2f, %.2f, %.2f'% (name, mse_cv.mean(), mse_cv.var(), mse_cv.std())
    print(mse)
```
